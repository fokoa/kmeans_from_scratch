```python
# Import
import numpy as np;
import pandas as pd;
from kmeans import KMeans as KM; # Our implementation of kmeans
from sklearn.decomposition import PCA;
import matplotlib.pyplot as plt;

%matplotlib inline
```

```python
# Load dataset
blobs = pd.read_csv('data/blobs.csv', delimiter=',');
```

```python
# Parameters
datasets = [
             (blobs, {'n_cluster':4, 'name':'blobs'})
           ]
```

```python
# Function to plot initial centroids
def plot(data, centroids_1, centroids_2, centroids_3):
   
    global j;
    centroids = [centroids_1, centroids_2, centroids_3];
    plt.figure(1, figsize=(12, 5), facecolor="w");
    
    for elt in centroids:
        plt.subplot(1, 3, j);
        plt.scatter(data[:, 0], data[:, 1], color='black', alpha=0.5, label='samples');
        for idx in range(0, elt.shape[0]):
            plt.scatter(elt[idx][0], elt[idx][1], color='red', label='centroids');
            if j == 1:
                plt.title('Initialisation avec random');
            elif j == 2:
                plt.title('Initialisation avec iter');
            elif j == 3:
                plt.title('Initialisation avec kmeans++');
        j = j + 1;
    
    plt.savefig('km_init.pdf', format='pdf');
    plt.show()
```

```python
global j;
j=1;

for data in datasets:
    if data[1]['name'] == 'iris':
        pca = PCA(n_components=2);
        X = pca.fit_transform(data[0][['a', 'b', 'c', 'd']].values);
        random_km = KM(n_clusters=data[1]['n_cluster'], init='random', n_init=1, seed=7);
        iter_km = KM(n_clusters=data[1]['n_cluster'], init='iter', n_init=1);
        kplus_km = KM(n_clusters=data[1]['n_cluster'], init='kmeans++', n_init=1, seed=7);
    else:
        X = data[0][['x', 'y']].values;
        random_km = KM(n_clusters=data[1]['n_cluster'], init='random', n_init=1, seed=7);
        iter_km = KM(n_clusters=data[1]['n_cluster'], init='iter', n_init=1);
        kplus_km = KM(n_clusters=data[1]['n_cluster'], init='kmeans++', n_init=1, seed=7);
    
    # Fit
    random_km.fit(X);
    iter_km.fit(X);
    kplus_km.fit(X);
    
    # Ploting
    plot(X, random_km.init_centroids[0], iter_km.init_centroids[0], kplus_km.init_centroids[0]);
    
```
